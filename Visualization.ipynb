### Use at least one of the clustering methods from the lecture. Clustering works best when the data are normalised (see Practical 8). Write a function to do the normalisation. Note that you usually want to show the original (not normalised values) to display the clustering results.

### Required packages for Elbow mthod

```python
# Required packages

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

### Elbow Method

```python
# Assigning parameter
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(2, 10)
```

```python
for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(Population)
    kmeanModel.fit(Population)
 
    distortions.append(sum(np.min(cdist(Population, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / Population.shape[0]) # Distortion List
    inertias.append(kmeanModel.inertia_) #Inertia list
 
    mapping1[k] = sum(np.min(cdist(Population, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / Population.shape[0]
    mapping2[k] = kmeanModel.inertia_
```

### Plotting Elbow Method graph using Inertia

```python
plt.plot(K, inertias, 'bx-') 
#Giving Axis Details
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('Elbow Method graph using Inertia')
plt.show()
```

Inference:

While look at the above Elbow method graph, we could say that number of clusters for population growth datast is 4

### Scaling the input data to fit the kmeans algorithm

```python
#Import scaler package for normalizing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(Population)
normal = scaler.transform(Population)
print("Normalized data is:", normal)
```

### Fitting K-Means Clustering Algorithm

```python
from sklearn.cluster import KMeans
#Fitting kmeans algorithm for n_clusters=4
model = KMeans(n_clusters=4)
model.fit(normal)
```

#'''Inverse Transformation'''

```python
# Doing inverse transformation
inverse_transformation = scaler.inverse_transform(normal) # Inverse scaling for the scaled data
print("Inversed Data", inverse_transformation)
```

Inference:
    The input datas are rescaled using inverse_transform method, After fitting kmeans model.

### Printing the clustering results

```python
plt.scatter(inverse_transformation[:,0],inverse_transformation[:,1], c=model.labels_, cmap='rainbow')
```

4 clusters are formed in population growth

### Scatter plot for un normalised data with centroid

```python
# Scatter plot for un normalised data with centroid
plt.scatter(inverse_transformation[:,0], inverse_transformation[:,1], c=model.labels_, cmap='rainbow')
plt.scatter(model.cluster_centers_[:,0] ,model.cluster_centers_[:,1], color='brown')
```

#### """Making low order polynomial for country Ireland"""

```python
"""Making low order polynomial for country Ireland"""
data=Population['Ireland']
f = 1/4

sine = np.sin(2*np.pi*f*data) + np.random.normal(size=len(data))
poly = np.polyfit(data, sine, deg=5) #Fitting the polynomial method

figure, axis = plt.subplots()
axis.plot(sine, label='data', color='red')
axis.plot(np.polyval(poly, data), label='fit', color='black')
plt.title('Low order polynomial curve fit')
axis.legend()
```

Low order polynomial curve fitted and graph drawn for non normalized data
